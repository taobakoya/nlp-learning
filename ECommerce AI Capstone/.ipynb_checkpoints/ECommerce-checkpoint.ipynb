```python
#create prepocessing the text - normalize, stopwords,lemmatilization..
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
wordnet_lemm=WordNetLemmatizer()
tokenizer=RegexpTokenizer(r'[a-z]+')
stop_words=set(stopwords.words('english'))
```

```python
def preprocess(document):
    document=document.lower()
    words=tokenizer.tokenize(document)
    words=[w for w in words if w not in stop_words]
    #lemmatization
    for pos in [wordnet.NOUN,wordnet.VERB,wordnet.ADJ,wordnet.ADV]:
        words=[wordnet_lemm.lemmatize(x,pos) for x in words]
    return " ".join(words)
print('done')
```

```python
data1['Processed_review']=data1['reviews.text'].apply(preprocess)
```

```python
data2=data1[['sentiment','Processed_review']]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidfData=TfidfTransformer().fit(reviews_bow)
tfidfDataFinal=tfidfData.transform(reviews_bow)
```

```python
# implement Naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB().fit(tfidfDataFinal,data2['sentiment'])
```

```python
final_data.head()
```

```python
#convert sentiment into numbers
df = final_data.copy()
```

```python
df.sentiment.replace(('Positive','Negative','Neutral'),(1,0,2),inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# prepare the data, split the data into train set & test set
X_train,X_test,y_train,y_test=train_test_split(df['Processed_review'],df['sentiment'],test_size=0.2,random_state=42)
```

```python
import tensorflow
```

```python
# import keras & tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Dropout,Embedding,GRU
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
top_words=20000
maxlen=100
nb_classes=3

# vectorize X_train and X_test to 2D tensor
tokenizer=Tokenizer(num_words=top_words)# consider only top 20000 words in the corpus
tokenizer.fit_on_texts(X_train)
```

```python
X_train_seq=tokenizer.texts_to_sequences(X_train)
X_test_seq=tokenizer.texts_to_sequences(X_test)
```

```python
from tensorflow.keras.preprocessing import sequence
```

```python
seq_train=sequence.pad_sequences(X_train_seq,maxlen=maxlen)
seq_test=sequence.pad_sequences(X_test_seq,maxlen=maxlen)
```

```python
seq_train.shape
```

```python
seq_train[:3]
```

```python
from keras.utils.np_utils import to_categorical
```

```python
# one hot encoding on y_train
y_train_seq=to_categorical(y_train,nb_classes)
y_test_seq=to_categorical(y_test,nb_classes)
```

```python
y_train_seq[:5]
```

# Construct a LSTM model
1. model building/define arch
2. compile the model
3. fit the model
4. make prediction

cat kitten-less than dog

```python
#model building LSTM
model1=Sequential()
model1.add(Embedding(top_words,128))
model1.add(Dropout(0.2))
model1.add(LSTM(128))
model1.add(Dropout(0.2))
model1.add(Dense(nb_classes,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# fit the model
model1.fit(seq_train,y_train_seq,
          batch_size=32,epochs=3)
```

```python
# model evaluation
score=model1.evaluate(seq_test,y_test_seq)
print('Test loss',score[0])
print('Test Accuracy',score[1])
```

```python
# model evaluation
score=model1.evaluate(seq_train,y_train_seq)
print('Train loss',score[0])
print('Train Accuracy',score[1])
```

```python
#model building GRU
model1=Sequential()
model1.add(Embedding(top_words,128))
model1.add(Dropout(0.2))
model1.add(GRU(128))
model1.add(Dropout(0.2))
model1.add(Dense(nb_classes,activation='softmax'))
```

```python
model1.summary()
```

```python
model1.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# fit the model
model1.fit(seq_train,y_train_seq,
          batch_size=32,epochs=3)
```

```python
# model evaluation
score=model1.evaluate(seq_test,y_test_seq)
print('Test loss',score[0])
print('Test Accuracy',score[1])
```

```python
# model evaluation
score=model1.evaluate(seq_train,y_train_seq)
print('Train loss',score[0])
print('Train Accuracy',score[1])
```

# Topic modeling
text - the phone is good, sound is good, camera quality

sports - cricket

```python
#create prepocessing the text - normalize, stopwords,lemmatilization..
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
wordnet_lemm=WordNetLemmatizer()
tokenizer=RegexpTokenizer(r'[a-z]+')
stop_words=set(stopwords.words('english'))
```

```python
def preprocess(document):
    document=document.lower()
    words=tokenizer.tokenize(document)
    words=[w for w in words if w not in stop_words]
    #lemmatization
    for pos in [wordnet.NOUN,wordnet.VERB,wordnet.ADJ,wordnet.ADV]:
        words=[wordnet_lemm.lemmatize(x,pos) for x in words]
    return " ".join(words)
print('done')
```

```python
doc_complete=df['Processed_review'].tolist()
doc_clean=[preprocess(doc).split() for doc in doc_complete]
```

```python
import gensim
from gensim import corpora
```

```python
dictionary=corpora.Dictionary(doc_clean)
```

```python
dictionary
```

```python
doc_term_matrix=[dictionary.doc2bow(doc) for doc in doc_clean]
doc_term_matrix
```

```python
from gensim.models import LdaModel
```

```python
NUM_TOPICS=9
ldamodel=LdaModel(doc_term_matrix,num_topics=NUM_TOPICS,id2word=dictionary,passes=30)
```

```python
topics=ldamodel.show_topics()
for topic in topics:
    print(topic)
    print()
```

```python
word_dict={}
for i in range(NUM_TOPICS):
    words=ldamodel.show_topic(i,topn=20)
    word_dict["Topic #"+"{}".format(i)]=[i[0] for i in words]
```
