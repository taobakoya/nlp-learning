<h3>Data Cleaning and processing</h3>

```python
import re
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
# ps = PorterStemmer()
wnl=WordNetLemmatizer()
corpus = []

for i in range(0, len(clndata)):
    review = re.sub('<.*>','',clndata['review'][i])
    review = re.sub('[^a-zA-Z]', ' ', clndata['review'][i])
    review = review.lower()
    review = review.split()
    
    # review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    #review = [wnl.lemmatize(word) for word in review if not word in stopwords.words('english')]
    
    review = ' '.join(review)
    corpus.append(review)
```
